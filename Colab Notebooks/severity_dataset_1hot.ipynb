```python
import numpy
import pandas as pd
import os
import torch
root2= '/gdrive/MyDrive/project/car-damage-dataset/data3a'
training = root2+'/training'
# we then split the training in train, valid so we store everything in the training folder
minor = '/01-minor/'
moderate = '/02-moderate/'
severe = '/03-severe/'

device = ("cuda" if torch.cuda.is_available() else "cpu")

train_df = pd.DataFrame(columns=["img_name","label"])
train_df_2 = pd.DataFrame(columns=["img_name","label"])
train_df_3 = pd.DataFrame(columns=["img_name","label"])




train_df["img_name"] = os.listdir(training+minor)
for idx, i in enumerate(os.listdir(training+minor)):
    train_df['label'][idx] = '1 0 0'
    # if idx>max_idx:
    #   max_idx = idx

# print(train_df)

train_df_2["img_name"] = os.listdir(training+moderate)
for idx, i in enumerate(os.listdir(training+moderate)):
    train_df_2['label'][idx] = '0 1 0'


train_df_3["img_name"] = os.listdir(training+severe)
for idx, i in enumerate(os.listdir(training+severe)):
    train_df_3['label'][idx] = '0 0 1'


# print(train_df_2)
frames = [train_df,train_df_2, train_df_3]
result_train_df = pd.concat(frames)
result_train_df=result_train_df.sample(frac=1).reset_index(drop=True)
print(result_train_df)
result_train_df.to_csv(r'/gdrive/MyDrive/project/'+'severity.csv', index = False, header=True)
```
